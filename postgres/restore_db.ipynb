```python
import psycopg2
import os

def restore_table(conn, table_file):
    """Restore data from a backup file to a PostgreSQL table."""
    with open(table_file, 'r') as f:
        with conn.cursor() as cursor:
            cursor.copy_expert(f"COPY \"{os.path.basename(table_file)[:-4]}\" FROM STDIN WITH CSV HEADER", f)

def restore_tables(database, user, password, host, port, backup_dir):
    """Restore all tables in a PostgreSQL database."""
    try:
        conn = psycopg2.connect(
            dbname=database,
            user=user,
            password=password,
            host=host,
            port=port
        )
        
        backup_files = [f for f in os.listdir(backup_dir) if f.endswith('.sql')]
        
        for table_file in backup_files:
            restore_table(conn, os.path.join(backup_dir, table_file))
        
        print("Restore completed successfully.")
    except psycopg2.Error as e:
        print("Error:", e)
    finally:
        if conn is not None:
            conn.close()
```

```python


if __name__ == "__main__":
    # Database connection parameters
    DATABASE = "vector_db"
    USER = "admin"
    PASSWORD = "admin"
    HOST = "localhost"
    PORT = "5433"

    # Directory to store backups
    BACKUP_DIR = "LLM\\_db_backup"

    # Perform restore
    restore_tables(DATABASE, USER, PASSWORD, HOST, PORT, BACKUP_DIR)
```
