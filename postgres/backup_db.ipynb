```python
import psycopg2
import os

def backup_table(conn, table_name, backup_dir):
    """Backup data from a PostgreSQL table to a file."""
    backup_file = os.path.join(backup_dir, f"{table_name}.sql")
    with open(backup_file, 'w') as f:
        with conn.cursor() as cursor:
            cursor.copy_expert(f"COPY \"{table_name}\" TO STDOUT WITH CSV HEADER", f)

def backup_tables(database, user, password, host, port, backup_dir):
    """Backup all tables in a PostgreSQL database."""
    try:
        conn = psycopg2.connect(
            dbname=database,
            user=user,
            password=password,
            host=host,
            port=port
        )
        with conn.cursor() as cursor:
            cursor.execute("SELECT table_name FROM information_schema.tables WHERE table_schema = 'public'")
            tables = [row[0] for row in cursor.fetchall()]
        
        for table_name in tables:
            backup_table(conn, table_name, backup_dir)
        
        print("Backup completed successfully.")
    except psycopg2.Error as e:
        print("Error:", e)
    finally:
        if conn is not None:
            conn.close()
```
